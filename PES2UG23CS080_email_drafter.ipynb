```python
from transformers import pipeline, set_seed
```

```python
generator = pipeline(
    "text-generation",
    model="distilgpt2"
)

set_seed(42)   # makes output more stable
```

```python
def generate_email(bullet_points):

    prompt = f"""
Bullet points:
Sick leave, Monday absence, fever, inform professor, submit assignment later

Email:
Dear Professor,

I hope you are doing well. I am writing to inform you that I was absent on Monday due to fever. I sincerely apologize for missing the class and will ensure that I submit the assignment as soon as possible.

Thank you for your understanding.

Regards,
Student


Bullet points:
{bullet_points}

Email:
Dear Professor,
"""

    result = generator(
        prompt,
        max_length=220,
        temperature=0.6,
        do_sample=True,
        repetition_penalty=1.3,
        pad_token_id=50256
    )

    generated = result[0]['generated_text']

    email_text = generated[len(prompt):]

    return "Dear Professor," + email_text
```

```python
print("Enter bullet points (comma separated):")

points = input()

email = generate_email(points)

print("\nGenerated Email:\n")
print(email)
```
