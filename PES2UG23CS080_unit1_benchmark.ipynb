```python
from transformers import pipeline
import torch

device = 0 if torch.cuda.is_available() else -1
print("Using device:", "GPU" if device == 0 else "CPU")
```

```python
bullet_points = "Sick leave, Monday absence, fever, inform professor, submit assignment later"

prompt = f"""
Bullet points:
{bullet_points}

Write a polite professional email.

Email:
Dear Professor,
"""
```

```python
models = {
    "DistilGPT2": "distilgpt2",
    "GPT2": "gpt2",
    "BART": "facebook/bart-base"
}
```

```python
for name, model in models.items():

    print(f"\n{name} Output:\n")

    try:
        generator = pipeline("text-generation", model=model, device=device)

        result = generator(
            prompt,
            max_length=200,
            temperature=0.7,
            repetition_penalty=1.3,
            pad_token_id=50256
        )

        generated = result[0]['generated_text']
        print(generated)

    except Exception as e:
        print("Failed:", e)
```

| Task | Model | Success/Failure | Observation | Technical Reason |
|------|-------|----------------|-------------|------------------|
| Email Auto-Drafting | DistilGPT2 | Partial Failure | Generated structured opening but content drifted away from task and ignored bullet points | Small autoregressive completion model without instruction tuning; prone to topic drift |
| Email Auto-Drafting | GPT2 | Partial Success | Produced fluent email-like text but did not follow provided context accurately | Larger language model improves fluency but lacks instruction-following capability |
| Email Auto-Drafting | BART | Failure | Output contained incoherent tokens and random text | Encoder-decoder architecture not suitable for text-generation pipeline without proper task setup |
